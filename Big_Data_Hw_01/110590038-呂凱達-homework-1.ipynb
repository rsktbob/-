````python
#data可以任意更動，data改變產生的檔案和輸出也會不同
data = '''葉聿欣,11112345,資工二,mary@gmail.com
李清蓁,11154246,智動四,fish@ntut.edu.tw
趙欣勇,11133250,電機三甲,brave@ntut.edu.tw
劉清,11161022,資財四甲,clear@ntut.edu.tw
歐陽素素,11155135,電機四丙,swing@ntut.edu.tw'''

data_lst = data.split('\n')
data_lst = [rows.split(',') for rows in data_lst]

data_column_length = [len(i) for i in data_lst[0]]

for rows in data_lst[1:]:
    for i in range(len(rows)):
        if data_column_length[i] < len(rows[i]):
            data_column_length[i] = len(rows[i])

data_chinese_lst = [False, False, False, False]

for i in range(len(data_lst[0])):
    if ord(data_lst[0][i][0]) > 128:
        data_column_length[i] *= 2
        data_chinese_lst[i] = True
    data_column_length[i] += 2

def print_table_line(data_length):
    for i in data_length:
        print('+'+'-'*(i), end='')
    print('+')

#印出資料
for rows in data_lst:
    print_table_line(data_column_length)
    for i in range(len(rows)):
        length = data_column_length[i]
        if data_chinese_lst[i]:
            element_length = len(rows[i])*2
        else:
            element_length = len(rows[i])
        if i == len(rows) - 1:
            print('| ' + rows[i] + ' '*(length-element_length - 1) + '|')
        else:
            print('|' + ' '*((length-element_length) // 2) + rows[i] + ' '*((length-element_length) // 2 + 1), end='')

print_table_line(data_column_length)

````

\begin{vmatrix}
a_{11} & a_{12} & \dots & a_{1n} \\
a_{21} & a_{22} & \dots & a_{2n} \\
\vdots & \cdots & \ddots & \vdots \\
a_{m1} & a_{m2} & \dots & a_{mn}
\end{vmatrix}$

$則 A_{m \times n}*B_{n \times p}=C_{m \times p}=[c_{ij}]_{m \times p}$ \
$$c_{ij} = \sum_{k=1}^{n}{a_{ik}b_{kj}}, \; for \: 1 <= i <= m, \: 1 <= j <= p$$

若已知矩陣
$A=\begin{vmatrix}
1 & 2 & 3 & 4\\
5 & 6 & 7 & 8\\
9 & 8 & 7 & 6
\end{vmatrix}_{3 \times 4}, 
B=\begin{vmatrix}
3 & 4 & 6\\

\end{vmatrix}_{4 \times 3},
C=\begin{vmatrix}
1 & 3 & 5 & 7\\

2 & 0 & 2 & 3
\end{vmatrix}_{3 \times 4}$, \
請計算下列矩陣運算式的結果:
1. $A+B$
2. $A+C$
3. $A*B$
4. $B*A$
5. $A*C$
6. $B*C$
8. $A*3$

````python
class Matrix:
    def __init__(self, row, col, values):
        # values: 型態是一維 list，裡面存的是矩陣元素的內容，順序為先列後行。 
        self.row = row
        self.col = col
        # 矩陣元素請以二維串列來儲存，請先完成下列二維串列(mat)的初值設定。
        self.mat = [[] for i in range(self.row)]
        for i in range(len(values)):
            self.mat[i // self.col].append(values[i])
        
    # 矩陣運算前，需要檢視行列是否符合計算規定！
    def __add__(self, other):
        if isinstance(other, Matrix):
            if self.row == other.row and self.col == other.col:
                values = []
                for rows in range(len(self.mat)):
                    for i in range(len(self.mat[rows])):
                        values.append(self.mat[rows][i]+other.mat[rows][i])
                return Matrix(self.row, self.col, values)
            else:
                return Matrix(0, 0, [])
        else:
            return Matrix(0, 0, [])

    def __sub__(self, other):
        if isinstance(other, Matrix):
            if self.row == other.row and self.col == other.col:
                values = []
                for rows in range(len(self.mat)):
                    for i in range(len(self.mat[rows])):
                        values.append(self.mat[rows][i]+other.mat[rows][i])
                return Matrix(self.row, self.col, values)
            else:
                print(False)
                return Matrix(0, 0, []) 
        else:
            print(False)
            return self
    
    # 乘法包含矩陣相乘，與矩陣與數值相乘。
    # 矩陣*矩陣, 矩陣*數值, 數值*矩陣，三種都要能進行運算！
    def __mul__(self, other):
        if isinstance(other, Matrix):
            if self.col == other.row:
                values = []
                
                def get_column(matrix, index):
                    return [matrix.mat[i][index] for i in range(matrix.row)]
                
                def vector_mul(m1, m2):
                    sum = 0
                    for i in range(len(m1)):
                        sum += m1[i]*m2[i]
                    return sum
                
                for i in range(self.row):
                    for j in range(other.col):
                        values.append(vector_mul(self.mat[i], get_column(other, j)))

                return Matrix(self.row, other.col, values)
            
            else:
                return Matrix(0, 0, [])
        else:
            values = []
            for rows in range(self.row):
                for i in range(self.col):
                    values.append(self.mat[rows][i]*other)
            return Matrix(self.row, self.col, values)

    def __rmul__(self, other):
            values = []
            for rows in range(self.row):
                for i in range(self.col):
                    values.append(self.mat[rows][i]*other)
            return Matrix(self.row, self.col, values)
    
    # 請為矩陣兩側加上'|'，原則上將元素當成三位數來列印。
    def __show__(self):
        if self.mat == []:
            print(False)
        for i in range(self.row):
            print('|', end='')
            for j in range(self.col):
                print(f'{self.mat[i][j]:<3}', end='')
            print('|')
        print()

A = Matrix(3, 4, [1, 2, 3, 4, 5, 6, 7, 8, 9, 8, 7, 6])
B = Matrix(4, 3, [3, 4, 6, 2, 5, 7, 6, 8, 9, 8, 4, 3])
C = Matrix(3, 4, [1, 3, 5, 7, 2, 4, 6, 8, 2, 0, 2, 3])

print('1')
(A + B).__show__()

print('2')
(A + C).__show__()

print('3')
(A * B).__show__()

print('4')
(B * A).__show__()

print('5')
(A * C).__show__()

print('6')
(B * C).__show__()

print('7')
(A * 3).__show__()

print('8')
(5 * B).__show__()
````
