```python
import pandas as pd

df = pd.read_csv('https://stats.moe.gov.tw/files/detail/111/111_student.csv')
df = df[['學校名稱', '總計']]
df['總計'] = df['總計'].str.replace(',', '')
df['總計'] = pd.to_numeric(df['總計'])
df = df.groupby('學校名稱').sum()

df = df.reset_index()

for i in df:
    print(i, end=' ')
print()

for i, j in zip(df['學校名稱'], df['總計']):
    print(i, j)
```

```python
import pandas as pd

df = pd.read_csv('https://stats.moe.gov.tw/files/detail/111/111_student.csv')
df = df[['學校名稱', '總計']]
df['總計'] = df['總計'].str.replace(',', '')
df['總計'] = pd.to_numeric(df['總計'])
df = df.groupby('學校名稱').sum()

df = df.reset_index()

df.to_html('111大專校院校別學生數.html', index=False)
```
